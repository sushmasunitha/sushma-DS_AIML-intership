# phase 3 -The Deep Dive (Univariate & Bivariate Analysis)

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("customer_analytics.csv")
df.head()
```

# PART 1: UNIVARIATE ANALYSIS
The histogram shows that most customers fall within the middle-age group. 
The customer base is concentrated between approximately 30 to 50 years.

```python
plt.figure()
plt.hist(df['Age'], bins=6)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Distribution of Customer Age")
plt.show()
```

The income distribution indicates variation in earning levels among customers. 
Some customers fall into higher income brackets, which may influence purchasing behavior.

```python
plt.figure()
plt.hist(df['AnnualIncome'], bins=6)
plt.xlabel("Annual Income")
plt.ylabel("Frequency")
plt.title("Distribution of Annual Income")
plt.show()
```

The bar chart shows the distribution of customers by gender. 
This helps understand whether the customer base is balanced or skewed toward one gender.

```python
plt.figure()
df['Gender'].value_counts().plot(kind='bar')
plt.xlabel("Gender")
plt.ylabel("Number of Customers")
plt.title("Gender Distribution")
plt.show()
```

# PART 2: BIVARIATE ANALYSIS
The scatter plot shows the relationship between income and spending behavior. 
Customers with higher income do not always have higher spending scores, indicating varied purchasing patterns.

```python
plt.figure()
plt.scatter(df['AnnualIncome'], df['SpendingScore'])
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Annual Income vs Spending Score")
plt.show()
```

The boxplot compares spending behavior between genders. 
It highlights differences in median spending scores and variation across groups.

```python
plt.figure()
sns.boxplot(x='Gender', y='SpendingScore', data=df)
plt.title("Spending Score by Gender")
plt.show()
```

# PHASE 4 – Customer Analytics Dataset
 The Big Picture (Multivariate Analysis & Storytelling)

```python
correlation_matrix = df.corr(numeric_only=True)
correlation_matrix
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure()
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

The correlation heatmap shows the strength and direction of relationships between numerical features.

SpendingScore shows a noticeable relationship with AnnualIncome and PurchaseFrequency, indicating that customer income and purchase behavior influence spending patterns.

OnlineVisitsPerMonth appears positively correlated with PurchaseFrequency, suggesting that customers who visit more frequently tend to purchase more often.

Age shows weaker correlation with spending behavior, meaning purchasing patterns are more behavior-driven than age-driven.

# Executive Summary

After performing multivariate analysis on the customer dataset, the following key insights were identified:

1. Customers with higher online visits tend to have higher purchase frequency, indicating strong digital engagement behavior.

2. SpendingScore is moderately influenced by AnnualIncome and PurchaseFrequency, suggesting that both earning capacity and buying habits impact customer value.

3. Age has a weaker relationship with spending patterns, indicating that behavioral factors are more important than demographic factors.

Overall, the dataset provides meaningful insights into customer purchasing behavior and can be further used for customer segmentation or predictive modeling.
