Task 1: The Paradigm Shift (Traditional vs Machine Learning)

Machine Learning is a method of teaching computers to learn patterns from data instead of explicitly programming rules. In traditional programming, we give the computer both data and rules to produce an output. However, in Machine Learning, we provide the computer with data and expected outputs, and it automatically learns the rules or patterns that connect them. This shift allows systems to improve their performance over time as they are exposed to more data. Machine Learning is widely used in applications such as spam detection, image recognition, recommendation systems, and self-driving cars.

Task 2: Three types of Machine Learning


1️. Supervised Learning

Data has input + labeled output
Model learns from examples

Real-world example:
Email Spam Detection,
Predicting house prices,
Classifying flower species (Our project)

2️. Unsupervised Learning

Data has no labeled output
Model finds patterns or groups by itself

Real-world example:
Customer segmentation,
Market basket analysis

3️. Reinforcement Learning

Model learns by trial and error
Gets rewards or penalties

Real-world example:
Self-driving cars,
Game-playing AI
